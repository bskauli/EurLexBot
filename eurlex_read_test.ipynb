```python
import requests
import re

from EUDirective import EUDirective

from bs4 import BeautifulSoup, Tag
```

```python
response = requests.get('https://eur-lex.europa.eu/legal-content/EN/TXT/HTML/?uri=CELEX:32014R0596')
```

```python
soup = BeautifulSoup(response.content.decode(), 'lxml-xml')
```

```python
date_regex = re.compile('\d{1,2}\s(January|February|March|April|May|June|July|August|September|October|November|December)\s\d{4}')
def text_contains_date(t):
    return re.search(date_regex, t)
    if re.match(date_regex, t):
        return True
    return False

def wrap_in_mark(s):
    return f"<mark>{s}<\mark>"

for element in soup.find_all('p', string=date_regex):
    wrapped = re.sub(date_regex, r'<mark>\g<0></mark>', element.text)
    print(wrapped)
    element.replace_with(BeautifulSoup(wrapped, 'html.parser'))

```
